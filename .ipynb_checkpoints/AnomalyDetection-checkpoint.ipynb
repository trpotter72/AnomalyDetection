# Anomaly Detection
We'll be looking through timeseries data to try and find anomalies.  To do this we'll be using isolation forests and local outlier factors.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
raw_data = pd.read_csv('./data/exchange-2_cpm_results.csv')
raw_data = raw_data.dropna()
raw_data.head()
```

```python
raw_data.info()
```

```python
#We'll want timestamps
raw_data["timestamp"] = pd.to_datetime(raw_data["timestamp"],infer_datetime_format=True)
raw_data["raw_timestamp"] = raw_data["timestamp"].map(datetime.timestamp)
raw_data = raw_data.dropna()
```

```python
plt.plot(raw_data["timestamp"], raw_data['value']) #50% of data between ~0.2 to ~0.5
plt.show()
```

```python
plt.boxplot(raw_data['value'])
plt.show()
```

## Lets add some features
A moving averages may help us decide when values fall out of range.  Before we add that, lets add in hour, and day of the week.

```python
eng = raw_data
eng["hour"] = eng["timestamp"].dt.hour
eng["day"] = eng["timestamp"].dt.dayofweek
eng['5_moving_avg'] = eng['value'].rolling(5).mean() 
eng['20_moving_avg'] = eng['value'].rolling(20).mean() 
eng['80_moving_avg'] = eng['value'].rolling(80).mean()
eng['400_moving_avg'] = eng['value'].rolling(400).mean()
```

```python
plt.plot(eng["timestamp"], eng[["5_moving_avg","20_moving_avg","80_moving_avg","400_moving_avg"]])
plt.show()
#This is some interesting data! Maybe we'll be able to find some cool info to help us out.
```

```python
#Lets also see how our std dev changes over time
eng['5_std_dev'] = eng['value'].rolling(5).std() 
eng['20_std_dev'] = eng['value'].rolling(20).std() 
eng['80_std_dev'] = eng['value'].rolling(80).std() 
eng['400_std_dev'] = eng['value'].rolling(400).std()
plt.plot(eng["timestamp"], eng[["5_std_dev","20_std_dev","80_std_dev","400_std_dev"]])
plt.show()
#Interestingly, the 20, 80, and 400 day std. dev. measurements are fairly regular, but the 5 hour measurements
#have their own high deviation
```

```python
eng.head()
```

```python
#Before we can detect anomalies, we'll have to drop NAN values and the timestamps col
eng = eng.dropna().drop('timestamp', axis = 1)
```

## Detecting Anomalies
Lets now create our models and find anomalies

```python
forest = IsolationForest(200)
labels = forest.fit_predict(eng)
eng.plot.scatter("raw_timestamp", "value", c = labels, cmap = 'Accent')
eng.plot.scatter("raw_timestamp","value") #this doesn't look great, let's try lowering contamination
```

```python
forest2 = IsolationForest(200,contamination = .05)
labels2 = forest2.fit_predict(eng.drop('raw_timestamp', axis = 1))
eng.plot.scatter("raw_timestamp", "value", c = labels2, cmap = 'Accent')
eng.plot.scatter("raw_timestamp","value") #this doesn't look great, let's try lowering contamination
```

### Still not looking great
Let's try doing it with less columns

```python
forest3 = IsolationForest(200)
labels3 = forest3.fit_predict(eng[["5_moving_avg","5_std_dev"]])
eng.plot.scatter("raw_timestamp", "value", c = labels3, cmap = 'Accent')
eng.plot.scatter("raw_timestamp","value") #This is starting to look good!
```

## Analysis
To see how we did with out isolation forest, let's try and plot with out the anomalies and see how that affects our means and std. devs.

```python
no_anoms = eng[labels == 1]
no_anoms.head()
```

```python
fig = plt.figure(figsize=(16,9))
ax2 = fig.add_subplot(122)
ax1 = fig.add_subplot(121, sharey = ax2)

ax2.plot(eng["raw_timestamp"], eng[["5_std_dev","20_std_dev","80_std_dev","400_std_dev"]])
ax1.plot(no_anoms["raw_timestamp"], no_anoms[["5_std_dev","20_std_dev",'80_std_dev','400_std_dev']])
plt.show()
```

The results above are promising.  We've been able to cut our max 5 hour std.dev in half.  It would be interesting to compare this data to the properly cleaned data to compare performance.  Just removing elements which contribute to a high standard deviation may not result in removing anomalies.
